# 1. Data Correction

## 1.1 Drop NA

```python
# tạo bảng df_dropna_all bằng cách loại bỏ tất cả giá trị null trong bảng và in ra info
df_dropna_all = df_transactions.dropna()
df_dropna_all.info()
```

```python
# tạo bảng df_dropna_online_order bằng cách loại bỏ giá trị null trong cột online_order và in ra info
df_dropna_online_order = df_transactions.dropna(subset = "online_order" )
df_dropna_online_order.info()
```

## 1.2 Fill NA

```python
# fill tất cả NA bằng 0, lưu vào bảng df_fillna_0 và in ra info
df_fillna_0 = df_transactions.fillna(0)
df_fillna_0.info()
```

```python
# fill NA bằng ffill, lưu vào bảng df_fillna_ffill và in ra info
df_fillna_ffill = df_transactions.fillna(method="ffill")
df_fillna_ffill.info()
```

```python
# fill NA bằng bfill, lưu vào bảng df_fillna_bfill và in ra info
df_fillna_bfill = df_transactions.fillna(method="bfill")
df_fillna_bfill.info()
```

## 1.3. Assess Categorical Data

```python
# load sheet Demographic và hiển thị top head
df_demographic = pd.read_excel('KPMG.xlsx', sheet_name='CustomerDemographic')
df_demographic.head()
```

```python
# in ra info bảng 
df_demographic.info()
```

```python
# describe các giá trị object
df_demographic.describe(include='object')
```

```python
# Remove all NA row
df_demographic = df_demographic.dropna()
```

```python
# In ra các giá trị unique của cột gender
print(df_demographic['gender'].unique())
# df_demographic_dropna.groupby("gender")["gender"].count()
```

```python
# Quy chuẩn cột Gender về 2 nhóm Male và Female, Unisex và in ra giá trị unique mới
# replace() syntax
df_demographic['gender'] = df_demographic['gender'].replace("F","Female")
df_demographic['gender'] = df_demographic['gender'].replace("M","Male")
df_demographic['gender'] = df_demographic['gender'].replace("Femal","Female")
df_demographic['gender'] = df_demographic['gender'].replace("U","Unisex")

print(df_demographic['gender'].unique())
```

```python
# Tính tuổi của mỗi khách hàng và lưu vào cột "Age" và in ra min age, max age
df_demographic['year_born'] = pd.DatetimeIndex(df_demographic['DOB']).year
now = datetime.now()
t = now.strftime("%Y")
df_demographic['current_year'] = float(t)
df_demographic["Age"] = df_demographic["current_year"] - df_demographic["year_born"]
df_demographic["Age"].agg([min, max])
# df_demographic["DOB"].agg([min, max])
```

```python
# phân nhóm khách hàng dưới 36 tuổi là nhóm "Young", Từ 36 - 55 là "Middle", trên 55 là Older
# lưu vào cột "Age_Group" bằng cách sử dụng hàm cut

df_demographic["Age_Group"] = pd.cut(x=df_demographic["Age"], bins=[1, 36, 55, 180],
                    labels=['Young', 'Middle', 'Older'])
df_demographic["Age_Group"] 
```

# 2. Data Completeness

```python
#join data sale vs demographic và in ra những trường hợp missing demographic info 
df_sales = pd.merge(df_fillna_bfill, df_demographic, how='inner', on = 'customer_id')
df_sales = df_sales.sort_values("transaction_id", ascending=True)
del df_sales['default']
# df_sales.to_csv("sales.csv")
print(df_sales.head())
print(df_sales.isnull().sum(axis = 0))
print(df_sales.isna().any(1))

# Bảng transactions là bảng fact nên khi join em cần giữ nguyên, nên ở đây how='left' nhé
```

# 3. Data Aggregation

```python
# tính tổng khách hàng bằng hàm nunique
total_of_customer = df_sales.nunique()['customer_id']
total_of_customer
```

```python
# Tính số khách khách hàng mỗi tháng
# hint 1: tạo column year_month bằng hàm dt.strftime
# hint 2: group by year_month và nunique

df_sales['year_month'] = df_sales['transaction_date'].dt.strftime("%Y-%m")
total_of_customer_by_month = df_sales.groupby('year_month', as_index=False).nunique()['customer_id']

total_of_customer_by_month
```

```python
# Tính Gross margin của mỗi tháng. Gross margin = list_price - standard_cost
# hint 1: tạo column Gross_Margin
# hint 2: group by year_month và sử dụng hàm agg
df_sales['Gross_Margin'] = df_sales['list_price'] - df_sales['standard_cost']
Gross_margin_by_month = df_sales.groupby('year_month', as_index=False)['Gross_Margin'].sum()
Gross_margin_by_month
```

```python
# Tính số lượng order và doanh thu group by online_order và order_status 
# hint 1: group by online_order, order_status và sử dụng hàm agg
order_type = df_sales.groupby(['online_order','order_status'], as_index=False).agg({
                    'transaction_id' : ['count'],
                    'list_price' : ['sum'] }).rename(columns={
                    'transaction_id': 'count_of_order','list_price': 'Total_revenue' })
order_type
```

# 4. Data Visualization

```python
# Sử dụng seaborn và vẽ ra 5 loại chart khác nhau (column, line, box,...) với đầy đủ tên chart, tên các trục và chú thích.
# Đưa ra message thông qua từng chart vừa vẽ
```

```python
#1 Total doanh thu, lợi nhuận theo từng Brand
_by_brand = df_sales.groupby(["brand"], as_index=False)[["list_price", "Gross_Margin"]].sum().sort_values("list_price",ascending=False)
_by_brand["Margin_rate"] =_by_brand["Gross_Margin"]/_by_brand["list_price"]*100
_by_brand
```

```python
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 6))

# Plot the total Revenue

sns.set_color_codes("pastel")
sns.barplot(x="list_price", 
            y="brand", 
            data=_by_brand,
            label="Revenue", 
            color="b")
# Plot the Profit
sns.set_color_codes("muted")
sns.barplot(x="Gross_Margin", 
            y="brand", 
            data=_by_brand,
            label="Profit", 
            color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(
       ylabel="",
       xlabel="Revenue and Profit by brand (million dollar)")
sns.despine(left=True, bottom=True)

plt.show()

# figsize anh thấy em chỉnh đang bất hợp lý, với horizontal bar thì kích thước nên đảo lại thành (15, 6) chart sẽ dễ nhìn hơn nhiều
```

```python
#2 doanh thu theo tháng
revenue_by_month = df_sales.groupby('year_month', as_index=False)['list_price'].sum().rename(columns={
                    'list_price': 'revenue'})
revenue_by_month
```

```python
sns.set_theme(style="whitegrid")

# Plot the trend of revenue by month
sns.lineplot(x="year_month", 
             y= 'revenue',
             marker=True,
             data=revenue_by_month).set(title='Revenue by month')
plt.xticks(rotation=45)
plt.show()

# rotation em nên set <= 45 thôi nhé, để 90 là mỗi lần nhìn trục x người ta lại phải nghiêng đầu 1 lần, mỏi cổ lắm :))
```

```python
#3 doanh thu theo tháng của từng dòng sản phẩm
revenue_by_month_by_product = df_sales.groupby(["product_line","year_month"], as_index=False)[["list_price"]].sum().sort_values(["product_line","year_month"],ascending=[True, True]).rename(columns={
                    'list_price': 'revenue'})
revenue_by_month_by_product
```

```python
sns.set_theme(style="whitegrid")

# Plot the trend of revenue by month by product_line
sns.relplot(data = revenue_by_month_by_product, 
               x ="year_month",
               y ="revenue",
               kind = "line",
               style = "product_line",
               markers = True).set(title='Revenue by month of product_line')
plt.xticks(rotation=45)
plt.show()

# comment về rotation tương tự chart bên trên :))
```
